```python
import pandas as pd
data = pd.read_csv("top_10_gdp_covid_data.csv")
```

```python
data.dtypes
```

```python
# This is the subset we are going to impute
num_data = data.select_dtypes(include='number')
num_data.dtypes
```

```python
# not going to be imputed
no_num = data.select_dtypes(include='object')
no_num.dtypes
```

```python
# Danling -> checked missing value percentage in their original dataset
# Mingyi -> used the same method to check the missing value percentage after combining the dataset
# After this round of selection, all our columns will have <20% missing value
# Making sure dataset is ready for imputation
def find_20_miss(df):
    missing_by_cln = df.isna().sum()
    percent_missing_by_cln = missing_by_cln/(df.shape[0])
    return percent_missing_by_cln.loc[percent_missing_by_cln <= 0.2]

def select_20_miss(df):
    selected_cln = list(find_20_miss(df).index)
    return df[selected_cln]
```

```python
num_data = select_20_miss(num_data)
# num_data = num_data.iloc[:,1:51]
num_data
```

```python
# Drop the column that's highly correlated to the target variable, drop the repetitive column
num_data = num_data.drop(['cumulative_confirmed','cumulative_deceased'], axis=1)
num_data
```

```python
# impute data
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn import impute
# Exclude the target variable when imputing
num_data_impute = num_data.drop('new_confirmed', axis=1)
imputer = impute.KNNImputer(n_neighbors=3)
imputer = imputer.fit(num_data_impute)
data_imputed = imputer.transform(num_data_impute)
```

```python
data_imputed = pd.DataFrame(data_imputed,columns = num_data_impute.columns)
```

```python
# Add the target varibale back
data_imputed['new_confirmed'] = num_data.iloc[:, 0]
# Add the object variables back
data_imputed = pd.concat([data_imputed, no_num], axis=1)
```
