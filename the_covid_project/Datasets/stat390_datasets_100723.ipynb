```python
useful_tables = [epidemiology, demographics, economy, emergency_declaration, geography, health, mobility, search_trends, vaccination,
        government_response, world_bank, key]
```

```python
# a function to see how many columns in each table is missing >20% of the data
def find_20_miss(df):
    missing_by_cln = df.isna().sum()
    percent_missing_by_cln = missing_by_cln/(df.shape[0])
    return percent_missing_by_cln.loc[percent_missing_by_cln <= 0.2]
```

```python
for table in useful_tables:
    print(find_20_miss(table))
```

Excluding World Bank table, other tables will make a aggregate table with ~40 columns. The World Bank table has about ~556 columns left. I decide to be more stringent with the World Bank data as I am only allow 10% missing.

```python
# a function to see how many columns in each table is missing >10% of the data
def find_10_miss(df):
    missing_by_cln = df.isna().sum()
    percent_missing_by_cln = missing_by_cln/(df.shape[0])
    return percent_missing_by_cln.loc[percent_missing_by_cln <= 0.1]
```

```python
print(find_10_miss(world_bank))
```

```python
list(find_20_miss(epidemiology).index)
```

```python
# select the columns with <20% missing
def select_20_miss(df):
    selected_cln = list(find_20_miss(df).index)
    return df[selected_cln]
```

```python
selected_epidemiology = select_20_miss(epidemiology)
selected_demographics = select_20_miss(demographics)
selected_economy = select_20_miss(economy)
selected_emergency_declaration = select_20_miss(emergency_declaration)
selected_geography = select_20_miss(geography)
selected_health = select_20_miss(health)
selected_mobility = select_20_miss(mobility)
selected_search_trends = select_20_miss(search_trends)
selected_vaccination = select_20_miss(vaccination)
selected_government_response = select_20_miss(government_response)
selected_key = select_20_miss(key)
```

```python
# for world bank, a 10% missing is allowed
def select_10_miss(df):
    selected_cln = list(find_10_miss(df).index)
    return df[selected_cln]
selected_world_bank = select_10_miss(world_bank)
```

```python
# concat tables together
agg_table1 = selected_epidemiology.merge(selected_emergency_declaration, on = ['date','location_key'], how = 'left')
```

```python
agg_table2 = agg_table1.merge(selected_demographics, on = ['location_key'], how = 'left')
```

```python
agg_table3 = agg_table2.merge(selected_geography, on = ['location_key'], how = 'left')
```

```python
agg_table4 = agg_table3.merge(selected_health, on = ['location_key'], how = 'left')
```

```python
agg_table5 = agg_table4.merge(selected_mobility, on = ['location_key', 'date'], how = 'left')
```

```python
agg_table6 = agg_table5.merge(selected_search_trends, on = ['location_key', 'date'], how = 'left')
```

```python
agg_table7 = agg_table6.merge(selected_government_response, on = ['location_key', 'date'], how = 'left')
```

```python
agg_table8 = agg_table7.merge(selected_key, on = ['location_key'], how = 'left')
```

```python
agg_table9 = agg_table8.merge(selected_vaccination, on = ['location_key', 'date'], how = 'left')
```

```python
agg_table10 = agg_table9.merge(selected_economy, left_on = ['country_code'], right_on = 'location_key', how = 'left')
```

```python
agg_table11 = agg_table10.merge(selected_world_bank, left_on = ['country_code'], right_on = 'location_key', how = 'left')
```
