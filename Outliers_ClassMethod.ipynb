```python
ax = x_train.new_deceased.plot(style='.',
                     figsize=(15,5),
                     color=color_pal[0],
                     title='new_deceased with bad data highlighted',
                     xlim=('2020-01-01','2021-12-31'),
                     ylim=(0, 3000))
x_train.query('new_deceased > 900').plot(style='.',
                                             figsize=(15,5),
                                             color=color_pal[3],
                                             ax=ax)
ax.legend().set_visible(False)
plt.show()
```

```python
# Clean Data
x_train = x_train.query('new_deceased < 900').copy()
```

```python
ax = x_train.new_persons_fully_vaccinated.plot(style='.',
                     figsize=(15,5),
                     color=color_pal[0],
                     title='new_persons_fully_vaccinated with bad data highlighted',
                     xlim=('2020-01-01','2021-12-31'),
                     ylim=(0, 900000))
x_train.query('new_persons_fully_vaccinated > 200000').plot(style='.',
                                             figsize=(15,5),
                                             color=color_pal[3],
                                             ax=ax)
ax.legend().set_visible(False)
plt.show()
```
